### Convert to CSV and  Change label as original dataset: -1 -> 1 and 1 -> 0

```python
import pandas as pd
file_path = 'diabetes_scale.txt'
with open(file_path, 'r') as file:
    lines = file.readlines()
data = []
for line in lines:
    parts = line.strip().split()
    label = parts[0]
    features = {int(f.split(':')[0]): float(f.split(':')[1]) for f in parts[1:]}
    features['label'] = label
    data.append(features)
df = pd.DataFrame(data)
sorted_columns = sorted([col for col in df.columns if isinstance(col, int)])
df = df[['label'] + sorted_columns]
df['label'] = df['label'].astype(int)
df['label'] = df['label'].replace({1: 0, -1: 1})
csv_file_path = 'Csv_File.csv'
df.to_csv(csv_file_path, index=False)
print(f"CSV file saved at: {csv_file_path}")
df.head(16)
```

#### KNN Imputer for missing values

```python
from sklearn.impute import KNNImputer
missing_values = df.isnull().sum()
imputer = KNNImputer(n_neighbors=5)
imputed_data = imputer.fit_transform(df.drop(columns=['label']))
df_imputed = pd.DataFrame(imputed_data, columns=df.columns[1:]) 
df_imputed['label'] = df['label'] 
df_imputed = df_imputed[['label'] + list(df.columns[1:])]
csv_imputed_file_path = 'Imputed_Csv_File.csv'
df_imputed.to_csv(csv_imputed_file_path, index=False)
print(f"Imputed CSV file saved at: {csv_imputed_file_path}")
df_imputed
df_imputed.head(15)
```
